# Model Card for GroupCW47-ModernBERT-NLI

<!-- Provide a quick summary of what the model is/does. -->

This is a fine-tuned version of answerdotai/ModernBERT-large designed for Natural Language Inference (NLI). It performs binary classification to determine whether a given hypothesis logically entails a premise.


## Model Details

### Model Description

<!-- Provide a longer summary of what this model is. -->

This model was developed for a university shared task on Natural Language Understanding. It takes pairs of sentences (a premise and a hypothesis) and predicts a binary relationship: 1 (ENTAILMENT) or 0 (NOT_ENTAILMENT). To handle the massive scale of the 395-million-parameter base model without exceeding standard GPU memory limits, it was trained using mixed precision (fp16) and gradient accumulation. Post-training, a dynamic probability threshold was applied to mathematically maximize final evaluation metrics.

- **Developed by:** CW47
- **Language(s):** English
- **Model type:** Transformer-based Sequence Classifier
- **Model architecture:** ModernBERT-Large (395M parameters)
- **Finetuned from model [optional]:** answerdotai/ModernBERT-large

### Model Resources

## Training Details

### Training Data

<!-- This is a short stub of information on the training data that was used, and documentation related to data pre-processing or additional filtering (if applicable). -->

The model was fine-tuned on a provided training dataset of 24,000 NLI sentence pairs (train.csv). Rows with NaN values were dropped, and labels were strictly mapped and cast to integers (0 and 1) to prevent device-side assertion failures on the GPU.

## Bias, Risks, and Limitations

<!-- This section is meant to convey both technical and sociotechnical limitations. -->

## Additional Information

<!-- Any other information that would be useful for other people to know. -->

Dynamic Threshold Optimization: Instead of relying on a standard 0.50 confidence cut-off, a custom loop was implemented post-training to extract the raw mathematical logits, convert them to probabilities via Softmax, and test all thresholds between 0.10 and 0.90 to scientifically maximize the target metric.
